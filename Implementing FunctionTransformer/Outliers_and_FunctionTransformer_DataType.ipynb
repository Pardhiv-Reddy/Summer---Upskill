DETECTING OUTLIERS
- HERE THE OUTLIERS ARE GENUINE AS THEY REPRESENT INFORMATION (RICH CUSTOMERS)

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
outliers = df[(df['Fare']<lower) | (df['Fare']>upper)]
print(outliers)
```

USING '.str.lower()' TO MAKE SURE THERE ARE NO INCONSISTENCIES

```python
df2.columns = df2.columns.str.lower()
df2.columns = df2.columns.str.strip()
```

```python
def date_decomp(X):
    X = X.copy()
    X['Year'] = X['date'].dt.year
    X['Month'] = X['date'].dt.month
    X['Day'] = X['date'].dt.day
    return X
def date_trans(X) :
    X = X.copy()
    X['date'] = pd.to_datetime(X['date'])
    return X
date_converter = FunctionTransformer(date_trans)
date_decomposer = FunctionTransformer(date_decomp)
```

```python
from sklearn.pipeline import Pipeline
date_pipeline  = Pipeline([
    ('to_date',date_converter),
    ('date_decomp',date_decomposer)
])
```

```python
print(df2['date'])
```

```python
df2 = date_pipeline.fit_transform(df2)
print(df2)
```
