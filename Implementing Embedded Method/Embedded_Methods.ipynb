```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression 
from sklearn.linear_model import Lasso
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin',axis=1,inplace = True)
```

```python
num_imp = SimpleImputer(strategy = 'median')
cat_imp = SimpleImputer(strategy = 'most_frequent')
df['Age'] = num_imp.fit_transform(df[['Age']])
df['Embarked'] = cat_imp.fit_transform(df[['Embarked']]).ravel()
```

```python
df.isnull().sum()
```

```python
scaler = StandardScaler()
df['Fare'] = scaler.fit_transform(df[['Fare']])
df['Age'] = scaler.fit_transform(df[['Age']])
df.head()
```

```python
X = df.drop('Survived',axis=1)
y = df['Survived']
```

Dropping Name and Ticket as they Increase Irrelevant Features in Encoding

```python
X.drop(['Name','Ticket'],axis=1,inplace = True)
X = pd.get_dummies(X)
```

```python
X.head()
```

```python
selector = SelectFromModel(LogisticRegression(l1_ratio=1,solver='saga',C = 0.05,max_iter = 10000))
X_selected = selector.fit_transform(X,y)
```

```python
X_selected_df = pd.DataFrame(X_selected, columns = selector.get_feature_names_out())
```
